```python
import pandas as pd
import numpy as np

dt1=pd.read_csv("C:/Users/shreya/Downloads/Data science Downloads/Python/project_2/carvan_train.csv")
dt2=pd.read_csv("C:/Users/shreya/Downloads/Data science Downloads/Python/project_2/carvan_test.csv")
```

```python
dt1.head()
```

```python
dt1.shape,dt2.shape
```

```python
dt1["data"]="train"
dt2["data"]="test"
dt2["V86"]=np.nan
```

```python
dt=pd.concat([dt1,dt2],0)
```

```python
dt.shape
```

```python
dt.head()
```

```python
var=["V4","V6","V44","V1","V5"]
for col in var:
    v=dt[col].value_counts()
    a=v.index[v>200][:-1]
    for ind in a:
        name = col +'_'+str(ind)
        dt[name] = (dt[col] == ind).astype(int)
    del dt[col] 
```

```python
dt.shape
```

```python
train=dt[dt["data"]=="train"]
train=train.drop("data",1)
test=dt[dt["data"]=="test"]
test=test.drop(["data","V86"],1)
```

```python
test=test.reset_index(drop=True)
```

```python
test.shape
```

```python
test
```

```python
x=train.drop("V86",1)
y=train["V86"]
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import roc_auc_score
```

```python
X_train,X_val,Y_train,Y_val=train_test_split(x,y,test_size=0.2,random_state=2)
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, roc_auc_score, fbeta_score, confusion_matrix
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import RandomizedSearchCV
from sklearn.metrics import make_scorer
```

```python
Y_train.value_counts(),Y_val.value_counts()
```

```python
param_dist = {"n_estimators":[500,450,550],
              "max_depth": [5, 10,15],
              "max_features":[55,60,65],
              "min_samples_split": [40,45,50],
              "min_samples_leaf": [25,30,35],
              "bootstrap": [True, False],
              "criterion": ["gini", "entropy"],
             "class_weight":["balanced"]}
n_iter_search=100
f_scorer = make_scorer(fbeta_score, beta=2)

random_search = RandomizedSearchCV(RandomForestClassifier(), param_distributions=param_dist,scoring = f_scorer,
                                   n_iter=n_iter_search,cv = 5, n_jobs = -1, verbose = 10)
```

```python
random_search.fit(X_train,Y_train)
```

```python
random_search.best_estimator_
```

```python
random_search.best_score_
```

```python
pred=random_search.predict(X_val)
```

```python
confusion_matrix(Y_val,pred)
```

```python
(pred == 0).sum(), (pred == 1).sum() 
```

```python
pred_final=random_search.predict(test)
```

```python
predictions=pd.DataFrame(pred_final)
```
